## Base functions for showing results

```python
import pandas as pd
from matplotlib import pyplot as plt

def create_diagram(optimizers: list, dataset: str, save_figure: bool, title: str, save_path: str):
    """
    Loads per-epoch metrics for multiple optimizers and plots validation loss vs. epochs.

    Reads `metrics/metrics_{optimizer}_{dataset}.csv` for each optimizer, expecting columns:
    [epochs, train_losses, train_accuracies, val_losses, val_accuracies, time_for_epoch].
    Aggregates all runs into one DataFrame, plots one line per optimizer, and optionally
    saves the figure as a PNG.

    Args:
        optimizers: List of optimizer names to compare (used in the metrics file names).
        dataset: Dataset identifier used in the metrics file names (e.g., "mnist", "cifar").
        save_figure: If True, saves the plot to `save_path`.
        title: Plot title.
        save_path: Output path for the PNG (only used if `save_figure` is True).
    """
    collected_data = pd.DataFrame(columns=["epochs", "train_losses", "train_accuracies", "val_losses", "val_accuracies", "optimizer"])

    for optimizer in optimizers:
        path = f"metrics/metrics_{optimizer}_{dataset}.csv"
        df = pd.read_csv(path, names=["epochs", "train_losses", "train_accuracies", "val_losses", "val_accuracies", "time_for_epoch"])
        df["optimizer"] = optimizer
        collected_data = pd.concat([collected_data, df], axis=0)

    figure = plt.figure(figsize=(8,8))
    for optimizer in optimizers:
        optimzer_data = collected_data[collected_data["optimizer"]==optimizer]
        plt.plot(optimzer_data["epochs"], optimzer_data["val_losses"], label = optimizer)

    plt.title(title)
    plt.xlabel("epochs")
    plt.ylabel("validation loss")
    plt.legend()
    if save_figure:
        figure.savefig(save_path, format="png")
    plt.show()
    
```

```python
import numpy as np

def create_diagram_convergence(optimizers: list, dataset: str, save_figure: bool, title: str, save_path: str):
    """
    Computes and visualizes convergence time per optimizer based on validation loss.

    For each optimizer, this function loads `metrics/metrics_{optimizer}_{dataset}.csv`
    with columns:
    [epochs, train_losses, train_accuracies, val_losses, val_accuracies, time_for_epoch].

    Convergence is defined as the epoch where `val_losses` reaches its minimum.
    The convergence time is the cumulative sum of `time_for_epoch` from epoch 1 up to
    (and including) that best epoch. Results are sorted by convergence time (fastest first)
    and plotted as a bar chart with one distinct color per optimizer and a seconds label
    above each bar.

    Args:
        optimizers: List of optimizer names to compare (used in the metrics file names).
        dataset: Dataset identifier used in the metrics file names (e.g., "mnist", "cifar").
        save_figure: If True, saves the plot to `save_path`.
        title: Plot title.
        save_path: Output path for the PNG (only used if `save_figure` is True).
    """
    collected_data = pd.DataFrame(columns=["epochs", "train_losses", "train_accuracies", "val_losses", "val_accuracies", "optimizer"])

    for optimizer in optimizers:
        path = f"metrics/metrics_{optimizer}_{dataset}.csv"
        df = pd.read_csv(path, names=["epochs", "train_losses", "train_accuracies", "val_losses", "val_accuracies", "time_for_epoch"])
        df["optimizer"] = optimizer
        collected_data = pd.concat([collected_data, df], axis=0)

    figure = plt.figure(figsize=(15,8))

    times = []
    epochs_to_converge = []

    for optimizer in optimizers:
        optimzer_data = collected_data[collected_data["optimizer"] == optimizer].copy()
        optimzer_data = optimzer_data.sort_values("epochs")

        idx = optimzer_data["val_losses"].astype(float).idxmin()
        epoch_to_converge = float(optimzer_data.loc[idx, "epochs"])
        time_until_convergence = float(optimzer_data.loc[optimzer_data["epochs"] <= epoch_to_converge, "time_for_epoch"].sum())

        times.append(time_until_convergence)
        epochs_to_converge.append(epoch_to_converge)

    # sort by convergence time
    sorted_idx = np.argsort(times)
    optimizers = [optimizers[i] for i in sorted_idx]
    times = [times[i] for i in sorted_idx]
    epochs_to_converge = [epochs_to_converge[i] for i in sorted_idx]

    # one color per optimizer
    cmap = plt.get_cmap("tab10")
    colors = [cmap(i % 10) for i in range(len(optimizers))]

    bars = plt.bar(optimizers, times, color=colors)
    plt.title(title)
    plt.xlabel("optimizer")
    plt.ylabel("time until best val loss (sec)")
    plt.xticks(rotation=25, ha="right")

    for bar, t, e in zip(bars, times, epochs_to_converge):
        plt.text(
            bar.get_x() + bar.get_width() / 2,
            bar.get_height(),
            f"{t:.2f}s",
            ha="center",
            va="bottom",
            fontsize=10
        )


    if save_figure:
        figure.savefig(save_path, format="png")
    plt.show()
```

```python
def create_diagram_total_runtime(optimizers: list, dataset: str, save_figure: bool, title: str, save_path: str):
    """
    Computes and visualizes total training runtime per optimizer across all epochs.

    For each optimizer, this function loads `metrics/metrics_{optimizer}_{dataset}.csv`
    with columns:
    [epochs, train_losses, train_accuracies, val_losses, val_accuracies, time_for_epoch].

    The total runtime is defined as the sum of `time_for_epoch` over all recorded epochs.
    Results are sorted by total runtime (fastest first) and plotted as a bar chart with
    one distinct color per optimizer and a seconds label above each bar.

    Args:
        optimizers: List of optimizer names to compare (used in the metrics file names).
        dataset: Dataset identifier used in the metrics file names (e.g., "mnist", "cifar").
        save_figure: If True, saves the plot to `save_path`.
        title: Plot title.
        save_path: Output path for the PNG (only used if `save_figure` is True).
    """
    collected_data = pd.DataFrame(columns=["epochs", "train_losses", "train_accuracies", "val_losses", "val_accuracies", "optimizer"])

    for optimizer in optimizers:
        path = f"metrics/metrics_{optimizer}_{dataset}.csv"
        df = pd.read_csv(path, names=["epochs", "train_losses", "train_accuracies", "val_losses", "val_accuracies", "time_for_epoch"])
        df["optimizer"] = optimizer
        collected_data = pd.concat([collected_data, df], axis=0)

    figure = plt.figure(figsize=(15,8))

    runtimes = []
    epochs_total = []

    for optimizer in optimizers:
        optimzer_data = collected_data[collected_data["optimizer"] == optimizer].copy()
        optimzer_data = optimzer_data.sort_values("epochs")

        runtime = float(optimzer_data["time_for_epoch"].astype(float).sum())
        runtimes.append(runtime)

        epochs_total.append(int(optimzer_data["epochs"].max()))

    # sort by total runtime (ascending)
    sorted_idx = np.argsort(runtimes)
    optimizers = [optimizers[i] for i in sorted_idx]
    runtimes = [runtimes[i] for i in sorted_idx]
    epochs_total = [epochs_total[i] for i in sorted_idx]

    # colors
    cmap = plt.get_cmap("tab10")
    colors = [cmap(i % 10) for i in range(len(optimizers))]

    bars = plt.bar(optimizers, runtimes, color=colors)

    plt.title(title)
    plt.xlabel("optimizer")
    plt.ylabel("total runtime (sec)")
    plt.xticks(rotation=25, ha="right")

    # labels on bars (seconds + epochs)
    for bar, rt, ep in zip(bars, runtimes, epochs_total):
        plt.text(
            bar.get_x() + bar.get_width() / 2,
            bar.get_height(),
            f"{rt:.2f}s",
            ha="center",
            va="bottom",
            fontsize=10
        )

    plt.tight_layout()

    if save_figure:
        figure.savefig(save_path, format="png", dpi=150, bbox_inches="tight")

    plt.show()

    
```

```python
from matplotlib.patches import Patch

def create_diagram_single_step_times(
    optimizers_to_experiment: list,
    csv_path: str,
    title: str,
    save_figure: bool,
    save_path: str,
    log_y: bool = False
):
    """
    Plots single-step (per-update) runtimes.

    Reads `csv_path` containing rows:
    [optimizer_index, diagThreshold, diagonalFlag, time_sec], maps optimizer_index to
    names via `optimizers_to_experiment`, and builds a bar chart sorted by runtime.

    For Shampoo variants ("shampoo", "shampoo_momentum", "shampoo_heuristic"), both
    diagonal and non-diagonal modes are plotted as separate bars (distinguished by
    hatching). For non-Shampoo optimizers, only the fastest measured time is shown.

    Args:
        optimizers_to_experiment: Optimizer names in the same order as used to write
            optimizer_index in the CSV.
        csv_path: Path to the timing CSV (no header).
        title: Plot title.
        save_figure: If True, saves the plot to `save_path`.
        save_path: Output path for the PNG (only used if `save_figure` is True).
        log_y: If True, uses a log-scaled y-axis for better visibility across
            large runtime differences.
    """
    # CSV: optimizer_index, diagThreshold, diagonalFlag, time_sec
    collected_data = pd.read_csv(
        csv_path,
        header=None,
        names=["optimizer_index", "diagThreshold", "diagonal", "time_sec"]
    )

    # index -> optimizer name
    idx_to_name = {i + 1: optimizers_to_experiment[i] for i in range(len(optimizers_to_experiment))}
    collected_data["optimizer"] = collected_data["optimizer_index"].map(idx_to_name)

    # diagonalFlag -> string
    collected_data["mode"] = np.where(collected_data["diagonal"] == 1, "diagonal", "non-diagonal")


    shampoo_variants = {"shampoo", "shampoo_momentum", "shampoo_heuristic"}

    bars = []  # (label, time_sec, base_optimizer, mode)
    for optimizer in optimizers_to_experiment:
        optimizer_data = collected_data[collected_data["optimizer"] == optimizer].copy()

        if optimizer in shampoo_variants:
            # both modi ((non)diagonal)
            for mode in ["diagonal", "non-diagonal"]:
                mode_data = optimizer_data[optimizer_data["mode"] == mode]
                if len(mode_data) == 0:
                    continue
                time_value = float(mode_data["time_sec"].min())
                label = f"{optimizer} {mode}"
                bars.append((label, time_value, optimizer, mode))
        else:
            # non-shampoo: take the fastest of both values
            time_value = float(optimizer_data["time_sec"].min())
            label = optimizer
            bars.append((label, time_value, optimizer, "best"))

    # sort from fast to slow
    bars.sort(key=lambda x: x[1])

    labels = [b[0] for b in bars]
    times = [b[1] for b in bars]
    base_opts = [b[2] for b in bars]
    modes = [b[3] for b in bars]

    figure = plt.figure(figsize=(15, 7))

    # per base optimizer one color
    cmap = plt.get_cmap("tab10")
    unique_base = []
    for b in base_opts:
        if b not in unique_base:
            unique_base.append(b)
    color_map = {b: cmap(i % 10) for i, b in enumerate(unique_base)}
    colors = [color_map[b] for b in base_opts]

    # Hatches: diagonal vs non-diagonal for Shampoo
    hatches = []
    for m in modes:
        if m == "diagonal":
            hatches.append("//")
        elif m == "non-diagonal":
            hatches.append("")
        else:
            hatches.append("")  # non-shampoo (best)

    bar_container = plt.bar(labels, times, color=colors)
    for rect, hatch in zip(bar_container, hatches):
        rect.set_hatch(hatch)

    plt.title(title)
    plt.xlabel("optimizer")
    plt.ylabel("time per update (sec)")
    if log_y:
        plt.yscale("log")
        plt.ylabel("time per update (sec, log scale)")

    # annotate seconds
    for rect, t in zip(bar_container, times):
        plt.text(
            rect.get_x() + rect.get_width()/2,
            rect.get_height(),
            f"{t:.2e}s",
            ha="center",
            va="bottom",
            fontsize=9
        )

    plt.xticks(rotation=25, ha="right")

    # legend for Hatches
    legend_handles = [
        Patch(facecolor="white", edgecolor="black", hatch="//", label="diagonal (Shampoo only)"),
        Patch(facecolor="white", edgecolor="black", hatch="", label="non-diagonal (Shampoo only)"),
        Patch(facecolor="white", edgecolor="black", hatch="", label="best-case (non-Shampoo)")
    ]
    plt.legend(handles=legend_handles, loc="upper left")

    plt.tight_layout()

    if save_figure:
        figure.savefig(save_path, dpi=150, bbox_inches="tight")

    plt.show()

```
